```python
# On importe les librairies dont on aura besoin pour ce tp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('house.csv')
```

```python
plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)
plt.show()
```

```python
# On décompose le dataset et on le transforme en matrices pour pouvoir effectuer notre calcul
X = np.matrix([np.ones(house_data.shape[0]), house_data['surface'].as_matrix()]).T
#print (X)
#print(house_data.shape[0])
y = np.matrix(house_data['loyer']).T
#print(house_data)
#print (y)
```

```python
# On effectue le calcul exact du paramètre theta
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
print(theta)
```

```python
plt.xlabel('Surface')
plt.ylabel('Loyer')

plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)

# On affiche la droite entre 0 et 250
plt.plot([0,250], [theta.item(0),theta.item(0) + 250 * theta.item(1)], linestyle='--', c='#000000')

plt.show()
```

```python
from sklearn import linear_model
import numpy as np
import pandas as pd
house_data = pd.read_csv('house.csv')
regr = linear_model.LinearRegression()
xx = np.matrix(house_data['surface']).T
yy = np.matrix(house_data['loyer']).T
#print(xx)
regr.fit(xx, yy)
#regr.predict(donnee_test)
m = regr.coef_[0]
b = regr.intercept_
print('b', b)
print('m', m)

plt.xlabel('Surface')
plt.ylabel('Loyer')

plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)

# On affiche la droite entre 0 et 250
plt.plot([0,250], [b,b + 250 * m], linestyle='--', c='#000000')

plt.show()
```
