```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('data.csv')
```

```python
plt.plot(house_data['area'], house_data['rent'], 'ro', markersize=4)
plt.show()
```

```python
#X is the area matrix : first column is 1 and second is area in m2
X = np.matrix([np.ones(house_data.shape[0]), house_data['area'].as_matrix()]).T
#print (X)
#print(house_data.shape[0])
y = np.matrix(house_data['rent']).T
#print(house_data)
#print (y)
```

```python
# we want the slope and the intercept of the regression
results = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
print(results)
print(results.item(0))
print(results.item(1))
```

```python
plt.title('MatPlot : Linear regression on a sample')
plt.xlabel('Area')
plt.ylabel('Rent')

plt.plot(house_data['area'], house_data['rent'], 'ro', markersize=4)

# display of the trend linear function
plt.plot([0,250], [results.item(0),results.item(0) + 250 * results.item(1)], linestyle='--', c='#000000')

plt.show()
```

```python
from sklearn import linear_model
import numpy as np
import pandas as pd
house_data = pd.read_csv('data.csv')
regr = linear_model.LinearRegression()
xx = np.matrix(house_data['area']).T
yy = np.matrix(house_data['rent']).T
#print(xx)
regr.fit(xx, yy)
#regr.predict(donnee_test)
m = regr.coef_[0]
b = regr.intercept_
print('b', b)
print('m', m)

plt.xlabel('Area')
plt.ylabel('Rent')

plt.plot(house_data['area'], house_data['rent'], 'ro', markersize=4)

# On affiche la droite entre 0 et 250
plt.plot([0,250], [b,b + 250 * m], linestyle='--', c='#000000')

plt.show()
```
